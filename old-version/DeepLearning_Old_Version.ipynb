```python
import nltk
nltk.download('punkt')

def preprocess(text):
    # Basic text cleaning
    if not isinstance(text, str):
        return ""
    return text.strip().replace('\n', ' ')

def split_sentences(text):
    return nltk.sent_tokenize(preprocess(text))

# Example: Preprocess candidate answers for extractive summary
# Iterate through the transposed DataFrame
candidates = []
for i, row in train_data_transposed.iterrows():
    # Access the 'answers' column for each row
    answers_dict = row.get('answers', {})
    if isinstance(answers_dict, dict):
        # Assuming answers_dict is a dictionary of answers for a single question
        for answer_key, answer_value in answers_dict.items():
            # Assuming answer_value is a dictionary containing 'answer_abs_summ' and 'answer_ext_summ'
            if isinstance(answer_value, dict):
                abstractive_summary = answer_value.get('answer_abs_summ', "")
                extractive_summary = answer_value.get('answer_ext_summ', "")
                candidates.extend(split_sentences(abstractive_summary))
                candidates.extend(split_sentences(extractive_summary))
            elif isinstance(answer_value, str):
                 candidates.extend(split_sentences(answer_value))

# Now 'candidates' contains sentences from all answers in the transposed data
# print(candidates[:10]) # Print first 10 sentences as an example
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

def extractive_summary(query, candidate_sentences, top_n=3):
    # Use TF-IDF similarity between question and candidate sentences
    vectorizer = TfidfVectorizer().fit([query] + candidate_sentences)
    q_vec = vectorizer.transform([query])
    c_vecs = vectorizer.transform(candidate_sentences)
    # Compute cosine similarities
    scores = np.asarray((c_vecs * q_vec.T).todense()).ravel()
    best_idxs = scores.argsort()[-top_n:][::-1]
    return [candidate_sentences[i] for i in best_idxs]

# Generate extractive summaries
single_extractive_sum = []
multi_extractive_sum = []

# Iterate through the transposed DataFrame to process each row
for i, row in train_data_transposed.iterrows():
    question = row['question']
    answers_dict = row.get('answers', {})

    # Extract sentences from all answer summaries for multi-extractive summary
    all_answer_sentences = []
    if isinstance(answers_dict, dict):
        for answer_key, answer_value in answers_dict.items():
            if isinstance(answer_value, dict):
                abstractive_summary_text = answer_value.get('answer_abs_summ', "")
                extractive_summary_text = answer_value.get('answer_ext_summ', "")
                all_answer_sentences.extend(split_sentences(abstractive_summary_text))
                all_answer_sentences.extend(split_sentences(extractive_summary_text))
            elif isinstance(answer_value, str):
                 all_answer_sentences.extend(split_sentences(answer_value))

    # Generate single extractive summary (using sentences from the first answer if available)
    first_answer_sentences = []
    if isinstance(answers_dict, dict) and len(answers_dict) > 0:
        first_answer_key = list(answers_dict.keys())[0]
        first_answer_value = answers_dict.get(first_answer_key, {})
        if isinstance(first_answer_value, dict):
            first_answer_sentences.extend(split_sentences(first_answer_value.get('answer_abs_summ', "")))
            first_answer_sentences.extend(split_sentences(first_answer_value.get('answer_ext_summ', "")))
        elif isinstance(first_answer_value, str):
            first_answer_sentences.extend(split_sentences(first_answer_value))


    if first_answer_sentences:
        single_extractive_sum.append(extractive_summary(question, first_answer_sentences, top_n=3))
    else:
        single_extractive_sum.append([])

    # Generate multi extractive summary (using sentences from all answers)
    if all_answer_sentences:
        multi_extractive_sum.append(extractive_summary(question, all_answer_sentences, top_n=5))
    else:
        multi_extractive_sum.append([])

# Now single_extractive_sum and multi_extractive_sum contain lists of summaries for each question
# You can print or process these lists as needed
# print("Single Extractive Summaries:", single_extractive_sum)
# print("Multi Extractive Summaries:", multi_extractive_sum)
```
